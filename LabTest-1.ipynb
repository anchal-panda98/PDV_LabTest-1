```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
states = pd.read_csv("Accident_States.csv")
cities = pd.read_csv("Accident_Cities.csv")
```

```python
states.fillna(states.mean())
cities.fillna(cities.mean())
```

```python
cities = cities.sort_values('Persons Injured - 2011', ascending = False)
```

```python
cities.head(3)[['Name of City', 'Persons Injured - 2011', 'Persons Injured - 2012', 'Injured - 2013', 'Injured - 2014', 'Injured - 2015']]
```

```python
states = states.sort_values('KIlled due to Over Speed', ascending = False)
```

```python
x = states.head(6)['Unnamed: 0']
y = ['2010','2011','2012','2013','2014','2015']
plt.bar(x,y)
plt.tight_layout()
plt.show()
```
