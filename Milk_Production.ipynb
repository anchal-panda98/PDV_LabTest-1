```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
milk_prod = pd.read_csv('milk_production(1).csv')
milk_prod.head()
```

```python
milk_prod.interpolate(method = 'linear', limit_direction = 'forward', inplace = True)
milk_prod
```
